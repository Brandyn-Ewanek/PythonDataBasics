## Conditionals if, else, elif

We will now look at conditional statements
There are 3 types of conditionals
1. if statement
2. else statement
3. elif statement

## if

```python
if True:
  print('true')
```

```python
if 1:
  print('true')
```

```python
if False:
  print('false')
```

```python
if 0:
  print('false')
```

```python
if a > b:
  print("{} is great than {}".format(a, b))
```

```python
if a < b:
  print("{} is less than {}".format(a, b))
```

```python
if a < b:
  print(f"{a} is less than {b}")
```

```python
if (a < b) & (a < d):
  print(f'{a} is less than {b} and {c}')
```

```python
if (a < b) & (a <= d):
  print(f'{a} is less than {b} and {c}')
```

## else

```python
if a > b:
  print(f"{a} is greater than {b}")
```

```python
if a > b:
  print(f"{a} is greater than {b}")
else:
  print(f"{a} is less than {b}")
```

## elif

```python
if a > d:
  print(f'{a} is greater than {d}')
elif a < d:
  print(f'{a} is less than {d}')
else:
  print(f'{a} is equal to {d}')
```

## one line if statement
