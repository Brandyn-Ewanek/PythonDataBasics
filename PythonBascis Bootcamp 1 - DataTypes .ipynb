## Python Slicing

selecting a secetion of data

inclusive : exclusive

```python
simple_list = ['data', 2, 3.7, [1,2,3], 'lists can hold anything']
print(simple_list[0])
print(simple_list[1])
print(simple_list[2])
print(simple_list[-1])
print(simple_list[3])
print(simple_list[3][0])
print(simple_list[1:4])
print(simple_list[1:])
print(simple_list[:-1])
print(simple_list[-1][10:15])
```

```python
# can insert into a list by indexing
simple_list[1] = 'new data'
print(simple_list)
print(simple_list[1])
print(simple_list[1][-4:])
```

# tuple

```python
# a tuple is an ordered collection of values that are unchangeable and allows duplicates
simple_tuple = ('data', 2, 3.7, [1,2,3], 'tuples can hold anything')
print(simple_tuple)
print(simple_tuple[1:4])
print(simple_tuple[1:])
print(simple_tuple[:4])
```

```python
# tuple
simple_tuple[1] = 'new data'
```

**Note**

Tuples are imutable but that is part of the reason they are faster than lists.  Used more in software engineering where speed matter, could be more important when putting a model into production.

# set

```python
# a set is an unordered collection of values that are changeable and does not allow duplicates
simple_set = {'data', 2, 3.7, [1,2,3], 'tuples can hold anything'}
```

```python
simple_set = {'data', 2, 3.7, 'tuples can hold anything'}
print(simple_set)
```

```python
print(simple_set[1])
```

```python
print('water' in simple_set)
```

```python
# you can't change values but you can add and remove entries from a set
simple_set.add(72)
print(simple_set)
simple_set.remove('data')
print(simple_set)
```

## DataFrame

Two dimenison square data structure.  Similar to SQL.

The spreadsheet of Python.  

Indices and columns allow use to access data. 


```python
# dictionary of lists (same length) fits well into a dataframe
# key is column title
# value is the column (one dimenion) of datatype

dict_df = {'col1':[6,1,2,3,3,3,4,5],
           'col2':[10.5, 2.5,3.25, 6.5, 7.5, 8.5, 9.5, 10.5],
           'col3':['one', 'two', 'three', 'three', 'three', 'six', 'seven', 'eight']
           }
pd.DataFrame(dict_df)
```

```python
df_simple = pd.DataFrame(dict_df)
```

When we select on column, note it changes into a Series.  No column title but keeps the indices.

```python
df_simple['col1']
```

```python
df_simple['col2']
```

```python
df_simple['col3']
```

Selecting multiple columns requires a list

```python
df_simple[['col1', 'col2']]
```

```python
df_simple[['col2', 'col1']]
```

Select rows with .loc or .iloc.  Location or index location.

Location is used when we've given new index values.

syntax: **.iloc[row:row, col:col]**



```python
df_simple.iloc[2]
```

```python
df_simple.iloc[2:5]
```

```python
df_simple.iloc[2: ]
```

```python
df_simple.iloc[ : , 1:]
```

### Change index values

```python
df_simple.index = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven']
df_simple
```

```python
df_simple.loc['zero']
```

```python
df_simple.loc[['two', 'four', 'zero'], ['col3', 'col1']]
```

### change column titles

```python
df_simple.columns = ['int', 'float', 'string']
df_simple
```
