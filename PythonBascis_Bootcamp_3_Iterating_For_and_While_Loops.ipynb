# For Loops

## Question 2

how many numbers are less than 50?

```python
less_50 = []

for num in list_nums:
  if num > 50:
    less_50.append(False)
  elif num < 50:
    less_50.append(True)
less_50
# sum(less_50) # True = 1, False = 0
```

## Question 3

how many numbers are greater than 200?

```python
great_200_true = []
great_200_false = []

for num in list_nums:
  if num > 200:
    great_200_true.append(True)
  elif num < 200:
    great_200_false.append(False)
print('number of false {}'.format(sum(great_200_false))) #len()
print('number of true {}'.format(sum(great_200_true)))
```

## Question 4

how many even numbers are there?

```python
even_num = []

for num in list_nums:
  if num % 2 == 0:
    even_num.append(1)

sum(even_num)
```

#  Filtering and Selecting Data

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/brandynewanek/data/main/dino.csv')
df.head()
```

## Question 10 Our measuring tool subtracted 2 from the height of all dino Decrease height of all dinos by two

```python
df['height'].mean()
```

```python
new_height = []
for row in df['height']:
  print(row)
  x = row + 2
  new_height.append(x)
```

```python
new_height
```

```python
df['new_height'] = new_height
df.head()
```

```python
df['new_height'].mean()
```

```python
ax = df['height'].plot(kind='hist', color='red', alpha=.7)
df['new_height'].plot(kind='hist', color='green', alpha=.7, title='new vs old height', ax=ax)
```

## Question 11 We've found the error was incorret by 2 for all types except subtracted 5 from type 3

```python
new_height2 = []
for i, row in df[['dino', 'height']].iterrows():
  print(row)
  if row[0] == 3:
    new_height2.append(row[1] + 5)
  
  else:
    new_height2.append(row[1] + 2)

df['new_height2'] = new_height2
```

```python
ax = df['height'].plot(kind='hist', color='red', alpha=.6)
ax = df['new_height'].plot(kind='hist', color='green', alpha=.6, ax=ax)
df['new_height2'].plot(kind='hist', color='purple', alpha=.4, title='new height2 vs firstcorrection vs old height', ax=ax)
```
